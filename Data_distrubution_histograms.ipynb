# Look at data

```python
import pandas as pd
df = pd.read_csv(path+'articles.csv')

df.head()
```

```python
categories = ['product_type_name', 'product_group_name', 'index_name', 'index_group_name', 'section_name', 'garment_group_name']

import matplotlib.pyplot as plt

for category in categories:
    plt.figure(figsize=(10, 6))
    df[category].value_counts().plot(kind='bar')
    plt.title(f"Counts of {category}")
    plt.xlabel(category)
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
